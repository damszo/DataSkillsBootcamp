```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Importing the data and listing first 5 rows to see where the age and charges columns are located.
my_data = pd.read_csv("insurance.csv")
my_data.head()
```

```python
# The iloc function allows use to select rows from the dataframe for plotting.  
x = my_data.iloc[:,:1].values
y = my_data.iloc[:,-1].values

plt.scatter(x,y,color = 'g')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Age vs. Charges')
plt.show()
```

```python
# Fitting the model
insurance_model = LinearRegression()
insurance_model.fit(x,y)
```

```python
# Plotting the data with best-fit line.
y_pred = insurance_model.predict(x)

plt.scatter(x,y,color = 'g')
plt.plot(x,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Age vs. Charges')
plt.show()
```

```python
# Predicting charges for someone at the age of 85.
unk_x = [[85]] 

x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = insurance_model.predict(x_pred)

plt.scatter(x,y,color = 'g')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Age vs. Charges')
plt.show()

print("Charges for 85 years old person should be:", insurance_model.predict(unk_x))
```
